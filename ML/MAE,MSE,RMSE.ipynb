```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load dataset
df = pd.read_csv('placement.csv')

# Visualize data
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package (in LPA)')
plt.title('CGPA vs Placement Package')
plt.show()

# Define features and target
x = df[['cgpa']]  # 2D array (n_samples, n_features)
y = df['package']  # 1D array

# Fit Linear Regression Model
model = LinearRegression()
model.fit(x, y)

# Predict using the model
y_pred = model.predict(x)

# Evaluation Metrics
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

# Adjusted R² score calculation
n = len(y)        # number of samples
k = x.shape[1]    # number of predictors (features)
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - k - 1)

# Print all metrics
print(f"Mean Squared Error (MSE): {mse:.3f}")
print(f"Mean Absolute Error (MAE): {mae:.3f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.3f}")
print(f"R² Score: {r2:.3f}")
print(f"Adjusted R² Score: {adjusted_r2:.3f}")

# Optional: Plot the regression line
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred, color='red', label='Predicted')
plt.xlabel('CGPA')
plt.ylabel('Package (in LPA)')
plt.legend()
plt.title('Linear Regression Fit')
plt.show()
```
