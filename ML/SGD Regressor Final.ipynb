```python
import numpy as np
import pandas as pd
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
df = pd.read_csv("experience_salary_dataset.csv")
```

```python
df.head(5)
```

```python
X = df[["experience_years"]].values
```

```python
y = df[["salary"]].values
```

```python
y
```

```python
sgd_reg = SGDRegressor(learning_rate="constant", eta0 = 0.01, max_iter = 1000)
```

```python
sgd_reg.fit(X,y)
```

```python
# Print final parameters
print(f"Final theta0 (intercept): {sgd_reg.intercept_[0]}")
print(f"Final theta1 (slope): {sgd_reg.coef_[0]}")
```
